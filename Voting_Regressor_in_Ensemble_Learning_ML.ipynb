# Regression :

```python
data = pd.read_csv("polynomial_reg.csv")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import r2_score

lr = LinearRegression()
rg = Ridge(alpha=0.3)
dt = DecisionTreeRegressor()

df = pd.DataFrame(data)
df.head()
x = df[["mileage"]]
y=  df['selling_price']

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)

vr = VotingRegressor(estimators=[("lr",lr),("rg",rg),("dt",dt)],weights=[2,1,1])    #WEIGHTS describes more weightage.

vr.fit(x_train,y_train)

y_prediction = vr.predict(x_test)

print("the r2_score report  : ")
r2_score  = r2_score(y_test,y_prediction)
print(r2_score)                                                   # r2 score is used for regression.
```
