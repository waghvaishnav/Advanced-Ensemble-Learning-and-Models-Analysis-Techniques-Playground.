```python
import pandas as pd

data = pd.read_excel("Raisin_Dataset  random forest.xlsx")

df = pd.DataFrame(data)
df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x =  df[["MajorAxisLength","MinorAxisLength","Eccentricity","ConvexArea","Extent","Perimeter"]]
y =  df["Area"]

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
```

```python
rf_model = RandomForestRegressor(n_estimators=50,random_state=42,max_features=7,max_depth=2,min_samples_split=3)
rf_model.fit(x_train,y_train)

y_predict = rf_model.predict(x_test)

mse =  mean_squared_error(y_test,y_predict)
print(mse)
# print("r2_score is ",r2_score(y_test,y_predict))
```
