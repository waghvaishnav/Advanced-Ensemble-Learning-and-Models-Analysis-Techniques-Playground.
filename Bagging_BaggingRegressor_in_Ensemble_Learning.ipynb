bagging using linear regression

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

data = load_diabetes()
x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

linear_model = LinearRegression()

bag_reg = BaggingRegressor(estimator=linear_model,n_estimators=21,random_state=42)

bag_reg.fit(x_train,y_train)

y_pred = bag_reg.predict(x_test)

print("the r2 score for th emodel is to be the :",r2_score(y_test,y_pred))
```

Bagging using ridge regression :

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import Ridge
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

data = load_diabetes()
x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

linear_model = Ridge()

bag_reg = BaggingRegressor(estimator=linear_model,n_estimators=21,random_state=42)

bag_reg.fit(x_train,y_train)

y_pred = bag_reg.predict(x_test)

print("the r2 score for th emodel is to be the :",r2_score(y_test,y_pred))
```
