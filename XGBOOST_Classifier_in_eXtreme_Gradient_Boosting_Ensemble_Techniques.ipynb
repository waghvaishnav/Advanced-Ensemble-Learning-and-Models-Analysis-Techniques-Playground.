```python
from xgboost import XGBClassifier
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier
import time

data = load_breast_cancer()
df = pd.DataFrame(data.data,columns=data.feature_names)
df["target"] = data.target

df.head()
```

```python
x = df.drop(columns=["target"])
y = df["target"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

# Training a Gradient Boosting Model :
gd_model = GradientBoostingClassifier()

start = time.time()
gd_model.fit(x_train,y_train)
gd_pred = gd_model.predict(x_test)
end = time.time()

r2 = r2_score(y_test, gd_pred)
mse = mean_squared_error(y_test, gd_pred)

print(f"R² Score: {r2:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print("Total time for the computation is ",end - start)      # it states that the gradient boost requires more computational time than xgboost.

print(classification_report(y_test,gd_pred))
```

```python
# Training a Xtreme Gradient Boosting Model :

xgb_model = XGBClassifier(n_estimator=100,learning_rate=0.1,random_state=42)

start1 = time.time()
xgb_model.fit(x_train,y_train)
xgb_pred = xgb_model.predict(x_test)
end1 = time.time()

r21 = r2_score(y_test, xgb_pred)
mse1 = mean_squared_error(y_test, xgb_pred)

print(f"R² Score: {r21:.4f}")
print(f"Mean Squared Error: {mse1:.4f}")
print("Total time for the computation is ",end1 - start1)      # it states that the gradient boost requires more computational time than xgboost.
```
