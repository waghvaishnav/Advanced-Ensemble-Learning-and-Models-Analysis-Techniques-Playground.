```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import classification_report
from sklearn.datasets import load_iris


lc = LogisticRegression()
sv = SVC(kernel="linear",probability=True)        # prob due to used voting soft.
dc = DecisionTreeClassifier()

data1 = load_iris()

x = data1.data
y=  data1.target

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)

vc = VotingClassifier(estimators=[("lc",lc),("sv",sv),("dc",dc)],voting="soft",weights=[2,1,1])   # voting used in classification only.hard / soft.

vc.fit(x_train,y_train)

y_prediction = vc.predict(x_test)

print("the classification_report  : ")
classification_report = classification_report(y_test,y_prediction)
print(classification_report)
```
